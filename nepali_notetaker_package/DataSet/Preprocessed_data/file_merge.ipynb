```python
import os
import pandas as pd
from glob import glob

# Path where your CSV files are stored
csv_folder = r'D:\nepali_notetaker_project\nepali_notetaker_package\DataSet\CSV_Data_file'  # <-- change this

# Get all CSV files in the folder
csv_files = glob(os.path.join(csv_folder, '*.csv'))
print(csv_files)
# Read and concatenate them
all_dfs = []

for file in csv_files:
    df = pd.read_csv(file, encoding='utf-8')  # use proper encoding
    all_dfs.append(df)

# Combine all into one DataFrame
combined_df = pd.concat(all_dfs, ignore_index=True)

# Save to new CSV
combined_df.to_csv('aaa_combined_dataset.csv', index=False, encoding='utf-8')
```
